However, despite the significant difference in performance of the model it is essential to acknowlegde the presence of outliers.

```python
def outliers(df, col="price"):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    ls = df.index[(df[col] < lower_bound) | (df[col] > upper_bound)]

    return ls.tolist()

outlier_list = outliers(df)
```

```python
df.drop(outlier_list, inplace=True)
print(f"The new shape of the dataset has {df.shape[0]} rows and {df.shape[1]} columns")
```

The new shape of the dataset has 530 rows and 13 columns
We can see that 15 outlier have been removed from the dataset which brings us to a new shape of 530 rows and 13 columns the removal of the outliers represents a 0.03% decrease in the size of the data which we can consider neglgible. On to training our model.

```python
X = df.drop(columns=["price"]).values.reshape(-1, 12)
y = df["price"].values
```

```python
# Assuming X and y are defined and imported earlier

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Scaling the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)  # Use transform instead of fit_transform for test data

# Implementing a multiple linear regression model
multiple_lr = LinearRegression()
multiple_lr.fit(X_train_scaled, y_train)
y_pred = multiple_lr.predict(X_test_scaled)

# Printing the metrics of the model
model_score = multiple_lr.score(X_test_scaled, y_test)
print(f"Model Score: {round(model_score * 100, 2)}%")

# Plotting the Actual vs Predicted Values as a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, color='blue', label='Actual vs Predicted')  # Scatter plot
plt.plot(y_test, y_test, color='red', label='Perfect Prediction')  # Plotting the perfect prediction line
plt.title("Actual vs Predicted Values (Scatter Plot)")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

From the model score we can see that even though our predictions are much better on the graph the score has decreased from 61.83% (Including Outliers) to 56.97% (Excluding Outliers). This could has happened due to the following factors

Loss of Information - The outliers may have contained valuable information about rare occurences or extreme cases in the data. Removing them lead to a loss of information which might have been essential for the model to generalize well to unseen data.

Model Sensitivity - Some models are more sensitive than others, thus removing outliers might have disrupted the assumptions of the model causing it to perform worse than before.

Overfitting - It is also possible that the model was overfitting to the outliers and removing them helped to alleviate overfitting.


```python
a=1
best_r2=-1

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    model=pipe.fit(X_train, y_train)
    y_pre=model.predict(X_test)

    print(f'{a} :- {name} - MSE: {mean_squared_error(y_test, y_pre):.4f}')
    print(f'     {name} - R2: {r2_score(y_test, y_pre):.4f}')
    print(f'     {name} - MAE: {mean_absolute_error(y_test, y_pre):.4f}',"\n")
    a+=1
    r2=r2_score(y_test, y_pre)
    if r2 > best_r2:
        best_r2 = r2
        best_model_name = name
        best_model = model
```

```python
print(f'The best model is: {best_model_name}')
print(f'The best model R2 Score is: {best_r2}')
```

```python
data_to_predict = [[7420, 4, 2, 3, 1, 0, 0, 0, 1, 2, 1, 0]]

if best_model is not None:
    prediction = best_model.predict(data_to_predict)
    print("Best Model:", best_model_name)
    print("Predicted Price:", prediction)
else:
    print("No best model found. Something went wrong.")
```
